```python
X = df.drop(['Target'], axis=1)

Y = df['Target']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x=scaler.fit_transform(X)
x
```

```python
y=Y
y
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier, XGBRFClassifier
from xgboost import plot_tree, plot_importance
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve
from sklearn import preprocessing
from sklearn.feature_selection import RFE
```

```python
#fitting different models and checking which performs best, note that i used same cell to test different models so some are not shown below
tr=GaussianNB()
tr.fit(x,y)
```

```python
p=tr.predict(x)
p
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score
print(score(p,y))
```

```python
xgbrf = XGBClassifier ( random_state=2021)
#xgbrf.fit(x,y)
```

```python
#performing grid search for hyper tuning xgbclassifier's parameters, turns out default params also perfomed well but i let it be

ss = {
    "n_estimators" : [100,200,500],
    "max_depth" : [3,6,9],
    "gamma" : [0.01,0.1],
    "learning_rate" : [0.001,0.01,0.1,1]
}
```

```python
from sklearn.model_selection import GridSearchCV
gs=GridSearchCV(estimator=xgbrf,
               param_grid=ss,
               scoring=["r2","neg_root_mean_squared_error"],
                refit="r2",
                cv=5,
                verbose=4)
gs.fit(x,y)
```

```python
print(gs.best_params_)
```

```python
print(gs.best_score_)
```

```python
xgbrf = XGBClassifier (gamma= 0.01, learning_rate= 0.1, max_depth= 9, n_estimators= 500)
xgbrf.fit(x,y)
```

```python
pgb=xgbrf.predict(x)
pgb
print(score(pgb,y))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,pgb)
```
