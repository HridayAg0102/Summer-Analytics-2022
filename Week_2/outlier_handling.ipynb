### `Outlier Analysis using Inner Quartile Range (IQR)`

video reference [LINK](https://youtu.be/A3gClkblXK8)

Example:


```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset = np.array([1.2, 2.3, 4.9, 5.1, 5.2, 5.4, 5.5, 5.5, 5.6, 5.6,
                    5.8, 5.9, 6, 6.1, 6.2, 6.5, 7.1, 8.8, 9.1, 10.9])
dataset = pd.DataFrame({"heights":dataset})
dataset
```

```python
sns.set_theme(rc = {'figure.figsize': (18, 5)} )
sns.set_style('white')
sns.boxplot(data = dataset, color = 'red', orient = "h")
```

```python
dataset.describe()
```

25 %ile ---> median of first 50% population
75 %ile ---> median of next 50% population

IQR = height(75%ile) - height(25%ile)

lower limit = height[25%ile] - 1.5* IQR

upper limit = height[75%ile] + 1.5* IQR

<font color = 'red'> NOTE: For calculating the 25%ile and 75%ile points, we exclude the median of the complete dataset and then consider the median of the rest of the observations in left and right halfs. </font>

Eg.

22 25 29 32 36 39 43 46 51

36 being the median must be excluded while calculating the points for 25%ile and 75%ile 

so points of 25ile = (25+29)/2
and 75%ile = (43+46)/2
