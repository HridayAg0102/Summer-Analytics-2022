# KDE plot (Kernal Density Plot)
used to find probability density function

````python
cars.head()
````

### Univariate (1 D)

````python
sns.kdeplot(cars.horsepower)
````

````python
sns.kdeplot(cars.horsepower, shade = True)
````

````python
sns.kdeplot(cars.horsepower, bw_method = 1, bw_adjust = .15)
````

Using the bw_method and bw_adjust, we generally vary the variance in the graph, so on reducing the values, we see more deviations in the peak and smoother plot on doing vice versa.

-[See from timestamp 4:02 to know what is bandwidth](https://youtube.com/clip/Ugkx6mAlefz7yw1tIC_ktWimt-mMrUFRGafs)

````python
sns.kdeplot(cars.horsepower, cumulative = True)

# To get Cumulative distribution function rather
# than Probability Density Function
````

### Bivariate KDE plot

````python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, legend = True)
````

````python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, shade = True)
````

````python
sns.kdeplot(x = cars.horsepower, y = cars.mpg, shade = True, cbar = True)
# Gives a color bar of density on adding cbar = True
````

# DisPlot

````python
sns.displot(x = 'weight', data = cars)
````

````python
# To add KDE plot on top
sns.displot(x = 'weight', data = cars, kde = True)
````

````python
# Kind Parameter gives different types of graph 
sns.displot(x = 'weight', data = cars, kind = 'ecdf')

#'ecdf' is a type of cumulative distribution plot
````

````python
sns.displot(x = 'weight', data = cars, kind = 'kde', rug = True)
# rug parameter gives lower pointers to graph for each datapoint
````

````python
# hue parameter seperates the curve in different categories.
# the categories are created according to the parameter passed in hue
sns.set_style('whitegrid')
sns.displot(x = 'weight', data = cars, hue = 'origin')

sns.displot(x = 'weight', data = cars, hue = 'origin', kind = 'kde')
````

````python
sns.displot(x = 'weight', \
data = cars, hue = 'origin', col = 'origin', rug = True)
````

````python
sns.displot(x = 'weight', data = cars, rug = True, rug_kws = {'height':0.09})
````

We can make several such changes in the graphs so see the docs for more info.

# HistPlot

````python
sns.__version__
````

````python
penguins = sns.load_dataset('penguins')
penguins.head()
````

````python
penguins.dropna(inplace = True)
````

````python
sns.histplot(x ='flipper_length_mm', data = penguins)
````

````python
sns.histplot(y ='flipper_length_mm', data = penguins)
````

````python
sns.histplot(x ='flipper_length_mm', data = penguins,binwidth = 10,\
binrange = (170,200))

#binrange fixes the range of the plotting
#binwidth fixes the width of the bars
#bins decide the number of bins in the graph

# We can have irregular bin sizes, i.e, 
#can fix the positions of the bars by doing
# bins = [170, 185, 200] i.e., passing the list of the bin locations.
````

### ```STATISTICS WITH HISTPLOT```

````python
sns.histplot(x ='flipper_length_mm',
             stat = 'density',
             fill = False,
             element = 'step',
             data = penguins)

# stat = density (area of histogram = 1),
# probability (sum of heights of bars = 1), count (default), etc...

# fill --> boolean
````

````python
sns.histplot(x ='flipper_length_mm',
             stat = 'density',
             fill = False,
             element = 'step',
             data = penguins,
            cumulative = True)
````

````python
sns.histplot(x ='flipper_length_mm',
             data = penguins,
             hue = 'species',
            multiple = 'fill')
````

## Byvariate Plot of Histogram

````python
sns.histplot(x ='bill_length_mm',y = 'species',
             data = penguins,
             hue = 'species',
             palette = "crest")

#palette is used for choosing color palette, cancarry a list of colors
#put a prebuilt palette name like crest , bone etc.
````

````python
sns.histplot(x ='bill_length_mm',y = 'bill_depth_mm',
             data = penguins,
             cbar = True)
````
